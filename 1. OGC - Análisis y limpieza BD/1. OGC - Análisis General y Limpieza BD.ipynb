```python
# FUNCIÓN PARA CREAR RANGOS EN FECHAS Y HORAS
def rango_dias(row):
    if row == 0: 
        return '0 a 24 horas'
    elif row == 1:
        return '1 a 2 días'
    elif row == 2:
        return '2 a 3 días'
    elif row == 3: 
        return '3 a 4 días'
    elif row == 4: 
        return '4 a 5 días'
    elif row == 5: 
        return '5 a 6 días'
    elif row == 6: 
        return '6 a 7 días'
    elif row>= 7 and row <= 14: 
        return '1 a 2 semanas'
    elif row>14 and row <= 21: 
        return '2 a 3 semanas'
    elif row>21 and row <= 31: 
        return '3 a 4 semanas'
    else:
        if row>0:
            return 'Más de 1 mes'
        else: 
            return 'Error en fechas'       
def rango_horas(row):
    if row == 0: 
        return 'Menos de 1h'
    elif row == 1:
        return '1 a 2 horas'
    elif row == 2:
        return '2 a 3 horas'
    elif row == 3: 
        return '3 a 4 horas'
    elif row == 4: 
        return '4 a 5 horas'
    elif row == 5: 
        return '5 a 6 horas'
    elif row == 6: 
        return '6 a 7 horas'
    elif row == 7: 
        return '7 a 8 horas'
    elif row>=8 and row <= 24: 
        return '8 a 24 horas'
    elif row>24 and row <= 72: 
        return '1 a 3 días'
    elif row>72 and row <= 168: 
        return '3 a 7 días'
    elif row>168 and row <= 730: 
        return '1 a 4 semanas'
    else:
        if row>0:
            return 'Más de 1 mes'
        else: 
            return 'Error en horas'
```

```python
# FUNCIÓN PARA MODIFICAR BASE DE DATOS, APLICANDO LA FUNCIONES ANTERIORES

def df_fechas_horas(df):
    d_meses = {1:'ENE',2:'FEB',3:'MAR',4:'ABR',5:'MAY',6:'JUN',7:'JUL',8:'AGO',9:'SEP',10:'OCT',11:'NOV',12:'DIC'}
    d_dias = {'Monday': 'Lunes','Tuesday':'Martes','Wednesday':'Miércoles','Thursday':'Jueves','Friday':'Viernes','Saturday':'Sábado','Sunday':'Domingo'}

    df['RECEPCION'] = pd.to_datetime((df['Recepción OT'] + ' ' +  df['Hora Recep.']), dayfirst= True)  
    df['ARRIBO'] = pd.to_datetime((df['Fecha Arribo'] + ' ' +  df['Hora Arribo']), dayfirst= True) 
    df['CIERRE'] = pd.to_datetime((df['Fecha Cierre'] + ' ' +  df['Hora Cierre']), dayfirst= True)

    df['Recepción OT'] = pd.to_datetime(df['Recepción OT'], dayfirst= True, format = '%d/%m/%Y')
    df['Fecha Arribo'] = pd.to_datetime(df['Fecha Arribo'], dayfirst= True, format = '%d/%m/%Y',errors='coerce')
    df['Fecha Cierre'] = pd.to_datetime(df['Fecha Cierre'], dayfirst= True, format = '%d/%m/%Y',errors='coerce')
    df['Hora Recep.']= pd.to_datetime(df['Hora Recep.'])
    df['Hora Arribo']= pd.to_datetime(df['Hora Arribo'],errors='coerce')
    df['Hora Cierre']= pd.to_datetime(df['Hora Cierre'],errors='coerce')

    df['Dia'] =  (df[['Recepción OT']].apply(lambda x: datetime.datetime.strftime(x['Recepción OT'], '%A'), axis=1)).apply(lambda x: d_dias[x])
    df['Nro Dia'] = df['Recepción OT'].dt.day
    df['Mes'] = (df['Recepción OT'].dt.month).apply(lambda x: d_meses[x]) +' '+(df[['Recepción OT']].apply(lambda x: datetime.datetime.strftime(x['Recepción OT'], '%y'), axis=1))
    df['Hora'] = df['Hora Recep.'].dt.hour

    def days_hours_minutes(td):
        return td.days, td.seconds//3600, (td.seconds//60)%60
    def hours_timestamp(td):
        from math import floor
        return (td.days*24)+(td.seconds//3600)

    df['Dias Recep/Arribo'] = (df['ARRIBO'] - df['RECEPCION']).dt.days
    df['Horas Arribo/Cierre'] = (df['CIERRE'] - df['ARRIBO'])
    df['Horas Arribo/Cierre'] = df['Horas Arribo/Cierre'].apply(hours_timestamp)

    df['Rango Recep/Arribo'] = df['Dias Recep/Arribo'].apply(rango_dias)
    df['Rango Arribo/Cierre'] = df['Horas Arribo/Cierre'].apply(rango_horas)
    return df
```

```python
# FUNCIÓN PARA VER LA EVOLUCIÓN DE LAS OT EN EL TIEMPO
def timeline_original(df, c = 'lightskyblue', figsize = [15,15]):
    plt.figure(1)
    #plt.suptitle("Timeline - Total OT",  fontweight="bold", fontdict={'fontsize':10})
    plt.rcParams['figure.figsize'] = figsize
    #print((f"\033[1mCantidad de meses: {len(sorted(df['Mes'].unique()))}\033[0m")) # Nro de meses
    meses = Counter(df["Mes"]) # Cantidad de mantenciones por mes
    #print(meses)
    plt.subplot(321)
    mi = ['JUL 18','AGO 18','SEP 18','OCT 18','NOV 18','DIC 18','ENE 19','FEB 19','MAR 19','ABR 19','MAY 19','JUN 19','JUL 19', 'AGO 19']
    mv = [meses[i] for i in mi]
    plt.plot(mi, mv, color = c, marker = 'o')
    plt.title("Cantidad de OT por mes", fontweight= 'bold', fontsize = 10)
    plt.ylabel('Número de OT')
    plt.xticks(rotation=30)
    #plt.ylim(1200,2000)

    # -----------------------------------------------------------
    plt.subplot(322)
    #print((f"\033[1mCantidad de días: {len(sorted(df['Dia'].unique()))}\033[0m")) # Nro de dias
    n_dias = Counter(df['Nro Dia']) # Cantidad de mantenciones por número de dias
    #print(n_dias)
    d4 = n_dias.most_common(31)
    ni = [i for i,v in d4][::-1]
    nv = [v for i,v in d4][::-1]
    #plt.rcParams['figure.figsize'] = figsize
    plt.bar(ni, nv, align='center', color = c)
    plt.title("Cantidad de OT por día",fontweight= 'bold', fontsize = 10)
    plt.ylabel('Número de OT')
    plt.xlabel('Dias')
    plt.xticks(range(1,32))

    # -----------------------------------------------------------
    plt.subplot(323)
    dias = Counter(df["Dia"]) # Cantidad de mantenciones por dia
    #print(dias)
    di = ['Lunes','Martes','Miércoles','Jueves','Viernes','Sábado','Domingo']
    dv = [dias[i] for i in di]
    #plt.rcParams['figure.figsize'] = figsize
    plt.plot(di, dv, color = c, marker = 'o')
    plt.title("Cantidad de OT por día semanal",fontweight= 'bold', fontsize = 10)
    plt.ylabel('Número de OT')
    plt.xticks(rotation=0)

    # -----------------------------------------------------------
    plt.subplot(324)
    #print((f"\033[1mCantidad de horas en Recepción OT: {len(sorted(df['Hora'].unique()))}\033[0m")) # Nro de horas
    horas = Counter(df["Hora"]) # Cantidad de mantenciones por rango de hora
    #print(horas)
    d3 = horas.most_common(30)
    hi = [i for i,v in d3][::-1]
    hv = [v for i,v in d3][::-1]
    #plt.rcParams['figure.figsize'] = figsize
    plt.bar(hi, hv, align='center', color = c)
    plt.title("Cantidad de OT por hora",fontweight= 'bold', fontsize = 10)
    plt.xlabel('Horas')
    plt.ylabel('Número de OT')
    plt.xticks(range(0,24))

    # -----------------------------------------------------------
    plt.subplot(325)
    #print((f"\033[1mDiferencia de tiempo entre la Recepción OT y el Arribo de Técnico: {len(sorted(df['Rango Recep/Arribo'].unique()))}\033[0m")) # Rango recep/arribo
    recep_arribo = Counter(df['Rango Recep/Arribo'])
    #print(recep_arribo)

    rango_di = ['0 a 24 horas', '1 a 2 días', '2 a 3 días','3 a 4 días','4 a 5 días','5 a 6 días','6 a 7 días', '1 a 2 semanas',
               '2 a 3 semanas','3 a 4 semanas','Más de 1 mes','Error en fechas']
    rango_dv = [recep_arribo[i] for i in rango_di]

    #plt.rcParams['figure.figsize'] = figsize
    plt.plot(rango_di, rango_dv, color = c, marker = 'o')
    plt.title("Tiempo entre recepción OT y arribo de técnico asignado",fontweight= 'bold', fontsize = 10)
    plt.ylabel('Número de OT')
    plt.xticks(rotation=50)

    # -----------------------------------------------------------
    plt.subplot(326)
    #print((f"\033[1mDiferencia de tiempo entre el Arribo y Cierre de trabajo: {len(sorted(df['Rango Arribo/Cierre'].unique()))}\033[0m")) # Rango arribo/cierre
    arribo_cierre = Counter(df['Rango Arribo/Cierre'])
    #print(arribo_cierre)

    rango_hi = ['Menos de 1h', '1 a 2 horas','2 a 3 horas','3 a 4 horas','4 a 5 horas','5 a 6 horas','6 a 7 horas',
              '7 a 8 horas','8 a 24 horas','1 a 3 días','3 a 7 días','1 a 4 semanas','Más de 1 mes','Error en horas']
    rango_hv = [arribo_cierre[i] for i in rango_hi]

    #plt.rcParams['figure.figsize'] = figsize
    plt.bar(rango_hi, rango_hv, color = c, align='center', width = 0.6)
    plt.title("Tiempo entre arribo y cierre de trabajo",fontweight= 'bold', fontsize = 10)
    plt.ylabel('Número de OT')
    plt.xticks(rotation=50)
    plt.tight_layout()
    #plt.subplots_adjust(top=0.92)
    #plt.savefig('timeline_original.jpg', dpi= 200, quality=95)
    plt.show()
```

# III. APLICACIÓN FUNCIÓN WORD_CLEANER() 

```python
# Limpieza de palabras
def word_cleaner(sent):
    import string,re
    from unidecode import unidecode
    text = sent.lower() 
    text = re.sub('\d{2}-\d{2}-\d{4}','',text) # Elimina fechas en formato DD-MM-YYYY
    text = re.sub('\d{2}/\d{2}/\d{4}','',text) # Elimina fechas en formato DD/MM/YYYY
    text = re.sub('\d{2}-\d{2}-\d{2}','',text) # Elimina fechas en formato DD-MM-YY
    text = re.sub('\d{2}/\d{2}/\d{2}','',text) # Elimina fechas en formato DD/MM/YY
    text = re.sub('\d-\d-\d{2}','',text)       # Elimina fechas en formato D-M-YY
    text = re.sub('\d/\d/\d{2}','',text)       # Elimina fechas en formato D/M/YY
    text = re.sub('\S*@\S*\s?','',text) # Elimina emails
    text = re.sub(r'\d{7,9}','', text) # Elimina teléfonos 
    text = re.sub('\[.*?¿\]\%', '', text) # Elimina .*?¿\
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text) # Elimina puntuación restante, como ¡! /
    text = re.sub('[%s]' % re.escape('°'), ' ', text)
    text = re.sub('[‘’“”…«»/]', '', text)
    text = re.sub('\n', ' ', text) # Elimina saltos de linea
    text = re.sub(r'\s{2,}', ' ', text) # Elimina dos o más espacios en blancos seguidos
    text = re.sub(r"^\s+", "", text, flags=re.UNICODE) # Elimina espacios blancos al principio
    text = re.sub("^\s+|\s+$", "", text, flags=re.UNICODE) # Elimina espacios blancos al final
    #text = re.sub('\w*\d\w*', '', text) # Elimina digitos
    #text = re.sub(r'(.)\1+', r'\1', text) # Elimina caracteres duplicados
    text = unidecode(text) # Elimina tildes y acentos (ñ -> n)
    return text 
```
